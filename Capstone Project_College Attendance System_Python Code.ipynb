```python
import mysql.connector
from datetime import date
```

```python
# ---------- Database Connection ----------

class Database:
    def __init__(self):
        self.conn = mysql.connector.connect(
            host="localhost",
            user="root",
            password="your_password",
            database="college_attendance"
        )
        self.cursor = self.conn.cursor()

    def commit(self):
        self.conn.commit()

    def close(self):
        self.cursor.close()
        self.conn.close()


# ---------- Student Class ----------

class Student:
    def __init__(self, db):
        self.db = db

    def add_student(self, name, course):
        query = "INSERT INTO students (name, course) VALUES (%s, %s)"
        self.db.cursor.execute(query, (name, course))
        self.db.commit()
        print("Student added successfully.")

    def view_students(self):
        query = "SELECT * FROM students"
        self.db.cursor.execute(query)
        for row in self.db.cursor.fetchall():
            print(row)


# ---------- Attendance Class ----------

class Attendance:
    def __init__(self, db):
        self.db = db

    def mark_attendance(self, student_id, status):
        query = """
        INSERT INTO attendance (student_id, attendance_date, status)
        VALUES (%s, %s, %s)
        """
        self.db.cursor.execute(query, (student_id, date.today(), status))
        self.db.commit()
        print("Attendance marked.")

    def view_attendance(self):
        query = """
        SELECT s.student_id, s.name, a.attendance_date, a.status
        FROM students s
        JOIN attendance a ON s.student_id = a.student_id
        """
        self.db.cursor.execute(query)
        for row in self.db.cursor.fetchall():
            print(row)


# ---------- Main Menu ----------

def main():
    db = Database()
    student = Student(db)
    attendance = Attendance(db)

    while True:
        print("\n--- College Attendance System ---")
        print("1. Add Student")
        print("2. View Students")
        print("3. Mark Attendance")
        print("4. View Attendance")
        print("5. Exit")

        choice = input("Enter choice: ")

        if choice == '1':
            name = input("Enter student name: ")
            course = input("Enter course: ")
            student.add_student(name, course)

        elif choice == '2':
            student.view_students()

        elif choice == '3':
            student_id = int(input("Enter student ID: "))
            status = input("Enter status (Present/Absent): ")
            attendance.mark_attendance(student_id, status)

        elif choice == '4':
            attendance.view_attendance()

        elif choice == '5':
            db.close()
            print("Exiting system...")
            break

        else:
            print("Invalid choice!")


if __name__ == "__main__":
    main()
```
