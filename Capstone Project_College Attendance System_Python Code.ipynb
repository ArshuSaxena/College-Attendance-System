```python

class Database:
    def __init__(self):
        self.connection = mysql.connector.connect(
            host = "localhost",
            user = "root",
            password = "devil",
            database = "college_attendance"
        )
        self.cursor = self.connection.cursor()

    def commit(self):
        self.connection.commit()

    def close(self):
        self.cursor.close()
        self.connection.close()
        
        
        
class Student:
    def __init__(self, db):
        self.db = db

    def add_student(self, name, course):
        query = "INSERT INTO students (name, course) VALUES (%s, %s)"
        self.db.cursor.execute(query, (name, course))
        self.db.commit()
        print("Student added successfully")

    def view_students(self):
        self.db.cursor.execute("SELECT * FROM students")
        for row in self.db.cursor.fetchall():
            print(row)


class Attendance:
    def __init__(self, db):
        self.db = db

    def mark_attendance(self, student_id, date, status):
        query = """
        INSERT INTO attendance (student_id, date, status)
        VALUES (%s, %s, %s)
        """
        self.db.cursor.execute(query, (student_id, date, status))
        self.db.commit()
        print("Attendance marked")

    def view_attendance(self):
        query = """
        SELECT students.name, attendance.date, attendance.status
        FROM attendance
        JOIN students ON attendance.student_id = students.student_id
        """
        self.db.cursor.execute(query)
        for row in self.db.cursor.fetchall():
            print(row)


from datetime import date

db = Database()
student = Student(db)
attendance = Attendance(db)

while True:
    print("\n--- College Attendance System ---")
    print("1. Add Student")
    print("2. View Students")
    print("3. Mark Attendance")
    print("4. View Attendance")
    print("5. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        name = input("Enter student name: ")
        course = input("Enter course: ")
        student.add_student(name, course)

    elif choice == "2":
        student.view_students()

    elif choice == "3":
        student_id = int(input("Enter student ID: "))
        status = input("Present / Absent: ")
        attendance.mark_attendance(student_id, date.today(), status)

    elif choice == "4":
        attendance.view_attendance()

    elif choice == "5":
        db.close()
        print("Goodbye!")
        break

    else:
        print("Invalid choice")
```
